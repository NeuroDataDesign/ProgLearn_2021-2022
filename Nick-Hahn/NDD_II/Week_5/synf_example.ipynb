```python
from proglearn import LifelongClassificationForest
from proglearn.sims import generate_gaussian_parity
import numpy as np
import matplotlib.pyplot as plt
```

```python
x,y = generate_gaussian_parity(10)
synf = LifelongClassificationForest()
synf.add_task(x,y,task_id=0)
test_x, test_y = generate_gaussian_parity(100)
errors = []

for i in range(100):
    x_,y_ = generate_gaussian_parity(10)
    synf.update_task(x_,y_,task_id =0)
    y_hat = synf.predict(test_x, task_id=0)
    errors.append(1-np.mean(y_hat==test_y))

plt.plot(np.linspace(10, 1000, 100),errors)
plt.ylabel("error")
plt.xlabel("Num Samples")
plt.title("Synf Streaming Error")
```
